```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x));
```

```python
plt.plot(x,np.sin(x));
```

```python
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x));
```

```python
plt.plot(x,np.sin(x-0),color='blue') #标准颜色名称
plt.plot(x,np.sin(x-1),color='g')  #缩写颜色代码（rgbcmyk）
plt.plot(x,np.sin(x-2),color='0.75') #范围在0-1的灰度值
plt.plot(x,np.sin(x-3),color='#FFDD44') #十六进制（RRGGBB,00-FF）
plt.plot(x,np.sin(x-4),color=(1,0,0.2,0.3)) #RGB元组，范围在0-1
plt.plot(x,np.sin(x-5),color='chartreuse'); #HTML颜色名称
```

```python
plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted');

#简写形式
plt.plot(x,x+0,linestyle='-') #实线
plt.plot(x,x+1,linestyle='--') #虚线
plt.plot(x,x+2,linestyle='-.') #点划线
plt.plot(x,x+3,linestyle=':'); #实点线
```

```python
plt.plot(x,np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5);
```

```python
plt.plot(x,np.sin(x))

plt.xlim(10,0)
plt.ylim(1.2,-1.2);
```

```python
plt.plot(x,np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x,np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x,np.sin(x))
plt.axis('equal');
```

```python
plt.plot(x,np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')

plt.legend();
```

```python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),
      xlabel='x',ylabel='sin(x)',
      title='A Sample Plot');
```
