```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

```python
ax1 = plt.axes() #默认坐标轴
ax2 = plt.axes([0.65,.65,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([.1,.5,.8,.4],
                  xticklabels=[],ylim=(-1.2,1.2))
ax2 = fig.add_axes([.1,.1,.8,.4],
                  ylim=(-1.2,1.2))

x = np.linspace(0,10)
ax1.plot(np.sin(x))
ax2.plot(np.cos(x));
```

```python
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(.5,.5,str((2,3,i)),
            fontsize=18,ha='center')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=.4,wspace=.4)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(.5,.5,str((2,3,i)),
            fontsize=18,ha='center')
```

```python
fig,ax = plt.subplots(2,3,sharex='col',sharey='row')
```

```python
#坐标轴存放在一个Numpy数组中，按照[row,col]取值
for i in range(2):
    for j in range(3):
        ax[i,j].text(.5,.5,str((i,j)),
                    fontsize=18,ha='center')
fig
```

```python
#创建一些正态分布数据
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean,cov,3000).T

#设置坐标轴和网格配置方式
fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4,hspace=.2,wspace=.2)
main_ax = fig.add_subplot(grid[:-1,1:])
y_hist = fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=main_ax)
x_hist = fig.add_subplot(grid[-1,1:],yticklabels=[],sharex=main_ax)

#主坐标轴画散点图
main_ax.plot(x,y,'ok',markersize=3,alpha=.2)

#次坐标轴画频次直方图
x_hist.hist(x,40,histtype='stepfilled',
           orientation='vertical',color='gray')
x_hist.invert_yaxis()

y_hist.hist(x,40,histtype='stepfilled',
           orientation='horizontal',color='gray')
y_hist.invert_xaxis()
```
